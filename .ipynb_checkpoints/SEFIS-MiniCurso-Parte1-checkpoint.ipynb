### O significado de Estatística!

Na ciência estamos interessados em interpretar as informações que conseguimos coletar da realidade em que vivemos, quaisquer que sejam esses dados. A Estatística é a ciência que nos auxilia nesta jornada, ela é uma ferramenta que metodologiza a forma de coletar, visualizar, manipular e interpretar as informações necessárias para um estudo científico.

Como qualquer ferramenta, ela deve ser utilizada com cuidado e conhecimento, para não causar erros interpretativos a cerca da realidade. Além disso, com prática, ela é cada vez mais uma grande aliada e facilitadora na extração de informações nas pesquisas!

### Dados

Para estudar a realidade, precisamos de informações. As informações podem ser extraídas através de medidas que inferimos. O conjunto de medidas da qual extraímos as informações é chamado de **dados**.

Os dados podem ser quantitativos (numéricos) ou qualitativos (não-numéricos). **Dados quantitativos** são medidas numéricas, ou seja, a qual as informações são escritas através de números. Já os **dados qualitativos**, são medidas que representam características (ou qualidades) do objeto de estudo, muitas vezes abstratas, assim sendo melhor descritas através de palavras.

Neste minicurso nos aprofundaremos nos dados quantitativos. Esses podem ser subdivididos ainda em 2 tipos: **discretos** (inteiros) e **contínuos** (reais). 

Exemplo de dados discretos: número de galáxias em um aglomerado, número de átomos de hidrogênio em uma molécula, etc.

Exemplo de dados contínuos: luminosidade de uma estrela, tempo de ocorrência de um processo físico, fluxo de energia, etc.

Quando selecionamos um conjunto de dados, na verdade estamos recortando um pedaço da realidade, ou seja, retirando uma **amostra** do que seria o conjunto de todos os dados possíveis, a **população**. É importante frisar que devemos ter cuidado e procurar retirar uma amostra representativa de toda a população e não apenas de uma parte, para que a nossa análise não seja comprometida.

A amostra ainda pode ser agrupada em conjuntos, ou "_**bins**_". Considerando dados discretos, os _bins_ podem ser formados por dados com valores iguais, já para dados continuos é selecionado um intervalo de valores a quais os dados de cada _bin_ pertence.

Esse agrupamento das amostras nos permite uma forma de visualizar dados quantitativos. O número de eventos em cada _bin_ nos permite construir **histogramas**. Nos histogramas a frequência de dados em um _bin_ é representada pela área de cada barra, então ele permite ilustrar como uma determinada amostra de dados distribuída.

.

**Exemplo:**

Conjunto de dados:

```python
#Bibliotecas para o funcionamento do código
import matplotlib.pyplot as plt
import numpy as np

#---

dados = np.random.normal(0.0, 0.5, 50) #Amostra de dados de uma população com distribuição normal (gaussiana). 
#dados
```

### Distribuições

Às medidas aleatórias pode haver associada uma probabilidade ou frequência de ocorrência, de forma que o conjunto de dados então pode ser descrito por uma **distribuição de probabilidade**. 

Exemplos:

**Gaussiana (ou Normal)**

$$
P (x; \mu, \sigma) = \frac{1}{\sqrt{ {\sigma}^2 2 \pi} } e ^ {- \frac{(x - \mu)^2 }{2 {\sigma}^2} },
$$

$\sigma$ é o desvio padrão e $\mu$ é a média.

```python
from scipy.stats import norm #Biblioteca para distribuição normal

plt.figure(figsize=(4,2.5)) 
plt.plot(np.linspace(-1,1), 
         norm(0.0, 0.5).pdf(np.linspace(-1,1))) #Gráfico da Normal com média 0.0 e desvio padrão 0.5.
plt.show()
```

**Uniforme**

$$
P (x; a, b) =
  \begin{cases}
      \frac{1}{ b - a }, \;\mbox{para}\; a \leq x \leq b. \\
      0, \;\mbox{em outro caso}\;. 
  \end{cases}  
$$

a e b são o ponto inicial e final do intervalo da distribuição.

```python
from scipy.stats import uniform #Biblioteca para distribuição uniforme

plt.figure(figsize=(4,2.5)) 
plt.plot(np.linspace(-1.5,2), 
         uniform(-0.5, 1.5).pdf(np.linspace(-1.5,2.0))) #Gráfico da uniforme com a = -0.5 e b = 1.0.
plt.show()
```

### Medidas de tendência central

Considerando um conjunto de N dados $x = \{x_1, x_2, ... ,x_N\}$ podemos verificar a localização "típica" dos dados considerando onde a maior parte deles se encontra. Isso é feito através das medidas de tendência central.


**Média aritmética**
$$
\bar{x} = \frac{1}{N} \sum\limits_{i=1}^{N} x_i.
$$

**Mediana**

É o valor do meio de um conjunto de dados.

**Moda**

É o valor mais popular em um conjunto de dados.

Exemplo:

```python
from statistics import median

dados_m = np.random.lognormal(0.1, 0.6, 100)
y, x, _ = plt.hist(dados_m, 10, color='aquamarine')

média_dados_m = dados_m.mean() #Média
moda_dados_m = (x[np.where(y == y.max())[0]] + x[np.where(y == y.max())[0] + 1]) / 2 #Moda
mediana_dados_m = median(dados_m) #Mediana

plt.axvline(moda_dados_m, color='k', label= "Moda")
plt.axvline(mediana_dados_m, color='magenta', label= "Mediana")
plt.axvline(média_dados_m, color='b', label= "Média")

plt.legend()
plt.show()
```

A média aritmética também pode ser calculada para funções e para os dados binados.

A média de uma função é:
$$
\bar{f} = \frac{1}{N} \sum\limits_{i}^{} f(x_i).
$$

A média para dados binados:
$$
\bar{x} = \frac{1}{N} \sum\limits_{i}^{} n_i x_i;
$$

$$
\bar{f} = \frac{1}{N} \sum\limits_{i}^{} n_i f(x_i).
$$
Em que $n_j$ é o número de elementos em cada bin.


**Outros tipos de média:**

Média geométrica:
$$
\bar{x}_g = \sqrt[N]{ x_1 + x_2 + ... + x_N }
$$

Média harmônica:
$$
\bar{x}_h = \frac{N}{1/x_1 + 1/x_2 + ... + 1/x_N}
$$

Média quadrática:

$$
\bar{x}_q =\sqrt{\frac{{x_1}^2 + {x_2}^2 + ... + {x_N}^2}{N}}
$$


### Medidas de dispersão

As medidas de dispersão fornecem o grau de variabilidade das medidas em relação a um valor central. Entre elas podemos destacar:

**Variância**

A variância é uma medida de dispersão sobre a média, ela mede o quanto os dados estão dispersos do valor médio.

$$
 Var (x) = \frac{1}{N} \sum\limits_{i}^{} (x_i - \bar{x})^2
$$

**Desvio padrão**

O desvio padrão é dado pela raiz da variância. Em uma distribuição gaussiana, ele mede a largura da distribuição.

$$
\sigma =  \sqrt{Var (x)} = \sqrt{ \frac{1}{N} \sum\limits_{i}^{} (x_i - \bar{x})^2}
$$

**Assimetria**


A assimetria mede o quão deslocados estão os dados para valores maiores ou menores. Quando a distribuição é negativa, ou enviesada a esquerda, os valores se concentram na extremidade maior da escala, quando a distribuição é positiva ou enviesada a direita, os valores se distribuem na extremidade menor.

$$
\gamma = \frac{1}{N \sigma^3} \sum\limits_{i}^{} (x_i - \bar{x})^3
$$

**Curtose**

Mede o grau de achatamento de uma distribuição dos dados.

$$
\kappa = \frac{1}{\sigma^4} \overline{ (x_i - \bar{x}) }^4 - 3
$$

Exemplos:

```python
import statsmodels.api as sm
from statistics import variance
from ipywidgets import interact
import ipywidgets as widgets
 
#Dados lineares

e=np.random.random_sample(20)
x= np.linspace(0, 10, num=20)
def linear_f(q):
    dt = 0.1 + 2*x + e*q
    
    var = variance(dt)
    res= sm.OLS(dt, sm.add_constant(x)).fit()
    plt.plot(x, res.fittedvalues, "k")
    plt.scatter(x, dt, c = 'b', label= f"Variância: {var}")
    plt.legend()
    plt.show()

interact(linear_f, q=widgets.FloatSlider(value=1, min=0.0, max=20, step=2));
```

```python
def plot_func(DP):
    plt.plot(np.linspace(-1,1), norm(0.0, DP).pdf(np.linspace(-1,1)), c='green') 
    plt.show()

interact(plot_func, DP=widgets.FloatSlider(value=0.5, min=0.1, max=1, step=0.1));
```

```python
from scipy.stats import beta

def beta_f(a,b):
    mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
    x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
    plt.plot(x, beta.pdf(x, a, b), 'magenta', label=f'Assimetria: {skew:.2f}; Curtose: {kurt:.2f}')
    plt.legend()
    plt.show()
    
interact(beta_f, a=widgets.FloatSlider(value=5, min=0.1, max=10, step=0.1), b=widgets.FloatSlider(value=5, min=0.1, max=10, step=0.1));
    
    
```

### Covariância e correlação

Muitas vezes ao longo de um estudo, não queremos obter informações de um conjunto de dados não apenas de uma espécie, mas de duas ou mais. Para medir o comportamento de um em relação ao outro podemos usar a covariância e a correlação.

**Covariância**

$$
cov(x,y) = \frac{1}{N} \sum\limits_{i}^{} (x_i - \bar{x})(y_i - \bar{y}) = \overline{xy} - \bar{x}\bar{y}
$$

**Correlação**

Coeficiente de correlação:
$$
\rho = \frac{cov(x,y)}{\sigma_x \sigma_y}
$$

Exemplos:

```python
import pandas as pd
from scipy import stats
import sklearn 
dados_planetas = pd.read_csv('~/Documentos/MESTRADO/SeFisica/planets_updated.csv', sep = ',')
#dados_planetas
```

### Erros

Quando fazemos estimativas através dos dados na maior parte das vezes obtemos apenas valores aproximados, isso se deve as incertezas. Assim, podemos estimar a diferença entre os valores estimados pelo modelo e os valores medidos através do cálculo dos erros.

Erro absoluto é o módulo da diferença entre os valores medidos $x_{med}$ e estimados $x_{aprox}$:

$$
e_{abs} = |x_{med} - x_{aprox}|
$$

Erro relativo:

$$
e_{rel} = \frac{|x_{med} - x_{aprox}|}{|x_{med}|}
$$

```python
d = np.array(dados_planetas['Diameter (km)'])
v = np.array(dados_planetas['Escape Velocity (km/s)'])
rl= sm.OLS(v, sm.add_constant(d)).fit()
plt.plot(d, rl.fittedvalues, "k")
plt.scatter(d, v, c = 'orange')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mean_absolute_percentage_error(v, rl.fittedvalues)
```
